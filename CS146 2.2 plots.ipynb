```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Normal distribtuon:
normal_1 = stats.norm(loc=0, scale=2)
normal_2 = stats.norm(loc=0, scale=1)
x = np.linspace(-5, 5)
plt.plot(x, normal_1.pdf(x), label = 'normal_1' )
plt.plot(x, normal_2.pdf(x), label = 'normal_2' )
plt.title("normal distribtuon")
plt.legend()
plt.show()
sample = normal_2.rvs(size = 100)
plt.hist(sample, density = True)
plt.plot(x, normal_2.pdf(x), label = 'normal_2' )
plt.title("normal dist and hist")
plt.show()
```

```python
#gamma distribtuon:
gamma_1 = stats.gamma(a=1, scale=1/2)
gamma_2 = stats.gamma(a=3, scale=1/3)
x = np.linspace(0, 5)
plt.plot(x, gamma_1.pdf(x), label = 'gamma_1' )
plt.plot(x, gamma_2.pdf(x) , label = 'gamma_2')
plt.title("gamma distribtuon")
plt.legend()
plt.show()

sample = gamma_1.rvs(size = 100)
plt.hist(sample, density = True)
plt.plot(x, gamma_1.pdf(x) )
plt.title("gamma dist and hist")
plt.show()
```

```python
#beta distribtuon:
beta_1 = stats.beta(a=2, b=1)
beta_2 = stats.beta(a=1, b=2)
x = np.linspace(0, 5)
plt.plot(x, beta_1.pdf(x) )
plt.plot(x, beta_2.pdf(x) )
plt.title("beta distribtuon")
plt.show()


sample = beta_1.rvs(size = 100)
plt.hist(sample,  density = True)
plt.plot(x, beta_1.pdf(x) )
plt.title("beta dist and hist")
plt.show()
```
